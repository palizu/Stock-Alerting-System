```python
import json 
import pandas as pd
import ssi_config as config
```

```python
df_yesterday =  pd.DataFrame(columns=['Symbol', 'TradingDate', 'Time', 'Open', 'High', 'Low', 'Close', 'Volume'])
with open("intra_data.json", 'r') as f:
    lines = f.readlines()
    for line in lines:
        data = json.loads(line)
        row = [data['Symbol'], data['TradingDate'], data['Time'], data['Open'], \
                data['High'], data['Low'], data['Close'], data['Volume']]
        df_yesterday.loc[len(df_yesterday)] = row
df_yesterday = df_yesterday.sort_values('Time')
df_yesterday
```

```python
df = pd.DataFrame(columns=['Symbol', 'TradingDate', 'Time', 'Open', 'High', 'Low', 'Close', 'Volume'])
with open("test_data.json", 'r') as f:
    lines = f.readlines()
    for line in lines:
        data = json.loads(line)
        row = [data['Symbol'], data['TradingDate'], data['Time'], data['Open'], \
                data['High'], data['Low'], data['Close'], data['Volume']]
        df.loc[len(df)] = row
df = df.sort_values('Time')
df
```

```python
res = pd.DataFrame(columns=df.columns)
for ticker in config.TICKERS:
    temp_df = df.loc[df["Symbol"] == ticker].copy()
    if len(temp_df) == 0:
        continue
    temp_df = temp_df.sort_values(by="Time")
    temp_df = temp_df.astype({"Volume": int, "Close":int})
    temp_df['TotalVol'] = temp_df['Volume'].cumsum()
    temp_df['Change'] = temp_df['Close'] - int(df_yesterday.loc[df_yesterday.Symbol == ticker, 'Close'].values[0])
    res = pd.concat([res, temp_df])
    
```

```python
res = res.astype({"Close":str, "Volume":str, "Change":str, "TotalVol":str})
res 
```

```python
with open("test_data.json", 'w') as f:
    for ind, row in res.iterrows():
        f.write(str(row.to_json()).replace("\\", ""))
        f.write("\n")
```
